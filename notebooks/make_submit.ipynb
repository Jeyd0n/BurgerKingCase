### Предскажим сначала значение колонки buy_post

```python
copy_tf_df = transformed_test.copy()
```

```python
copy_tf_df
```

```python
copy_tf_df.drop(['customer_id', 'startdatetime'], axis=1, inplace=True)
```

```python
copy_tf_df
```

```python
import pickle

path = '../model/new_random_forest_classifier.pickle'

classifier = pickle.load(open(path, "rb"))
```

```python
buy_post = classifier.predict(
    X=copy_tf_df
)
```

```python
buy_post = pd.Series(buy_post, name='buy_post')
```

```python
user_ids = user_ids.reset_index(drop=True)
```

```python
user_ids
```

```python
buy_post
```

```python
half_submit = pd.concat((user_ids, buy_post), axis=1)
```

```python
half_submit
```

```python
path = '../model/random_forest_regressor.pickle'

regressor = pickle.load(open(path, "rb"))
```

```python
date_diff_post = regressor.predict(
    X=copy_tf_df
)
```

```python
date_diff_post = pd.Series(date_diff_post, name='date_diff_post')
```

```python
date_diff_post = round(date_diff_post, 0)
```

```python
half_submit = pd.concat((user_ids, date_diff_post), axis=1)
```

```python
submit = pd.concat((half_submit, buy_post), axis=1)
```
